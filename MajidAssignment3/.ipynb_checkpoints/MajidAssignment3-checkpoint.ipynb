# Q6. Which year is the coldest during the winter season? Show all the year minimum temperature!

```python
#There are 2 ways to solve the question. 
#First Way - Showing the year with minimum temperature in WINTER
#Second Way - Showing the year with minimum mean temperature in WINTER
#I will solve the question by both ways..................
```

# Q6 First Way

```python
#Showing all year min temperature

weathercolumn = weather.loc[weather['season'] == 'Winter']
grouped5 = weathercolumn.groupby(['year'])
functions2 = ['min']
result2 = grouped5['tempC'].agg(functions2)
result2
```

```python
#Which year was the coldest in Winter?
resultswint = result2.sort_values(ascending=True, by=['min'])
resultswint
mintempyear = resultswint[:1].ix[:,['min']]
mintempyear
```

```python
#Hence Year 2013 was the coldest year in Winter
```

# Q6 2nd Way

```python
#Which year was the coldest in Winter? - the one with lowest mean temp. in winters
weathercolumn = weather.loc[weather['season'] == 'Winter']
grouped5 = weathercolumn.groupby(['year'])
functions2 = ['mean']
result2 = grouped5['tempC'].agg(functions2)
resultswint = result2.sort_values(ascending=True, by=['mean'])
mintempyear = resultswint[:1].ix[:,['mean']]
mintempyear
```

```python
#Again using the second method, 2003 was having the lowest mean temperature in Winters
```

# Q7. Which year is the hottest during the summer season? Show all the year maximum temperature!

```python
#For question 7, lets solve using just one method. i.e
# Year with higest mean temperature in Summer is the hotest YEAR
```

```python
#Showing the year which is hottest during the summer season
weathercolumn2 = weather.loc[weather['season'] == 'Summer']
groupedsummer = weathercolumn2.groupby(['year'])
functions3 = ['mean']
resultsummer = groupedsummer['tempC'].agg(functions3)
resultssummer = resultsummer.sort_values(ascending=False, by=['mean'])
maxtempyear = resultssummer[:1].ix[:,['mean']]
maxtempyear
```

```python
#So 2010 is the year with highest mean temperature in Summer
```

```python
#Showing all year's max temperatures
weathercolumn2 = weather.loc[weather['season'] == 'Summer']
groupedsummer = weathercolumn2.groupby(['year'])
functions3 = ['max']
resultsummer = groupedsummer['tempC'].agg(functions3)
resultsummer
```

# Q8. Are there any temperature differences between odd and even year? Show your argument! (use the annual average temperature)

```python
#Calculating Mean of Means of Even Years
evenyear = weather[(weather['year'] % 2 == 0)]
evengrouped = evenyear.groupby(['year'])
evengroupedmean = DataFrame(evengrouped['tempC'].agg('mean'))
evengroupedmean['tempC'].mean()
```

```python
#Calculating Mean of Means of Odd Years
oddyear = weather[(weather['year'] % 2 != 0)]
oddgrouped = oddyear.groupby(['year'])
oddgroupedmean = DataFrame(oddgrouped['tempC'].agg('mean'))
oddgroupedmean['tempC'].mean()
```

```python
#Difference of both the means
```

```python
print ("Temperature Difference b/w Even and Odd year is:" 
       + str(evengroupedmean['tempC'].mean() - oddgroupedmean['tempC'].mean()))
```
