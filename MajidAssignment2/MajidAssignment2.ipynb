# Q2. Combining all 3 .dat files to a DataFrame

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat', sep='::', header=None, names=mnames)
#showing all the movies 
#movies[:]
```

```python
unames = ['user_id', 'gender','age','occupation', 'zip']
users = pd.read_table('users.dat', sep='::', header=None, names=unames)
#showing all the usersdata 
#users[:]
```

```python
rnames = ['user_id', 'movie_id','rating','timestamp']
ratings = pd.read_table('ratings.dat', sep='::', header=None, names=rnames)
#showing all the usersdata 
#ratings[:]
```

```python
moviesdf = DataFrame(movies)
```

```python
ratingsdf = DataFrame(ratings)
```

```python
usersdf = DataFrame(users)
```

```python
dfmerged = moviesdf.merge(ratingsdf, how='inner', on='movie_id').merge(usersdf, on='user_id',how='inner')
```

```python
def binsfunct(agevalue):
    
    return ('Young' if (agevalue <= 10) 
            else 'Youth' if(11 <= agevalue <= 20) 
            else 'Adult' if (21 <= agevalue <= 65) 
            else 'Senior')
```
