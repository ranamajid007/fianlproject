# Fixing Embarked Missing Value

```python
#For Embarked we can fill it with most frequent embarked location. Lets find out most used embarked location.
```

```python
ascdfprofile = dfprofile['Embarked'].sort_values(ascending=True)
ascdfprofile[:1]
```

# Fixing Age Missing Values

```python
#As we can see that age has missing values i.e ouf of 418 passengers (417 boarded) we have AGE Data for just 332 passengers
#There is 2 ways to handle this data
#First Method: Drop and remove the AGE Column fromt the entire dataset
#Second Method: Add Age Median to all the missing values
#Since we have age data for most of the passengers, I would prefer not to drop the column, rather add the AGE Median to 
#all the missing values
#I did not choose Mean since the results will be in decimal points which doesn't make sence for age
```

```python
dfprofile['Age'].median()
```

```python
#Filling in the missing values in the Age column with the Median i.e 27
dfprofile['Age'].fillna(dfprofile['Age'].median(), inplace=True)
dfprofile.info()
```

```python
#Since we still have a Fare column with NaN value we can put a mean value in it. 
```

# Fixing Fare missing value

```python
#Filling in the missing values in the Fare column with the Mean
dfprofile['Fare'].fillna(dfprofile['Fare'].mean(), inplace=True)
dfprofile.info()
```

# Cleaning Ticket Column

```python
#Ticket column is not used throughout our analysis. We can either drop it or can clean it by our choice
#I cleaned it by keeping only the First Part of tickets before any character or spacegap
```

```python
#REF: https://www.kaggle.com/shivendra91/rolling-in-the-deep
def ticketcleaning(ticket):
        ticket = ticket.replace('.','')
        ticket = ticket.replace('/','')
        ticket = ticket.split()
        ticket = map(lambda t : t.strip() , ticket)
        ticket = filter(lambda t : not t.isdigit(), ticket)
        ticket = list(ticket)
        if (len(ticket)) > 0:
            return ticket[0]
        else: 
            return 'ABC'

dfcombined['Ticket'] = dfcombined['Ticket'].map(ticketcleaning)
dfcombined
```

# Age Role on Survival Rate

```python
#First let us calculate total survivros and non-survivors
```

```python
totalsurv = dfcombined[dfcombined['Survived']==1]
totalnosurv = dfcombined[dfcombined['Survived']==0]

print("Total Survived are: %i (%.1f percent), Total Dead are: %i (%.1f percent), Total Passengers: %i"\
      %(len(totalsurv), 1.*len(totalsurv)/len(dfcombined)*100.0,\
        len(totalnosurv), 1.*len(totalnosurv)/len(dfcombined)*100.0, len(dfcombined)))
```

```python
#creating new column for Baby, Young, Adult and Old
def binsfunct(agevalue):
    
    return ('Child' if (agevalue <= 10) 
            else 'Young' if(11 <= agevalue <= 20) 
            else 'Adult' if (21 <= agevalue <= 50) 
            else 'Old')

dfcombined['AgeClass'] = dfcombined['Age'].map(binsfunct)
```

```python
#Let us calculate total number of survivors and non-survivors based on these 4 AgeClasses
```

```python
age_grouped = dfcombined.groupby(['AgeClass','Survived'])
agegroupstats = age_grouped.describe()
functions = ['count']
result = age_grouped['Survived'].agg(functions)
result.sort_values(by = 'count', ascending=False)
```

```python
#From the above analysis it looks like no. of survived adults is highest i.e 97 but this is not useful.
#Lets calculate mean of the survivors per AgeClass to get better idea.
```

```python
age_grouped = dfcombined.groupby(['AgeClass'])
agegroupstats = age_grouped.describe()
functions = ['mean']
result = age_grouped['Survived'].agg(functions)
result.sort_values(by = 'mean', ascending=False)
```

```python
#So according to my distributaion, Old (age more than 51) had 0.48% of survival.
#While child and adull have less probabiliyt to survive
```

```python
#It can also be proved by the graphs
```

```python
plt.close('all')
fig = plt.figure()
plt.hist([dfcombined[dfcombined['Survived']==1]['AgeGroup'],dfcombined[dfcombined['Survived']==0]['AgeGroup']], color = ['g','r'],
         label = ['Survived','Dead'])
plt.xlabel('AgeGroup')
plt.ylabel('Number of passengers')
plt.legend()
plt.show()
```

# Grouping on the basis of Fair

```python
#Let us first find out min, max and mean value of the Fair
dfFare = dfcombined.describe()
dfFare['Fare']
#So Minimum Fare is 0
# Maximum Fare is 512.32
```

```python
#Lets make Fair classes based on 50,50 steps
```

```python
#creating new column for Baby, Young, Adult and Old
def binsfunct(farevalue):
    
    return ('ExtremeLow' if (farevalue <= 50) 
            else 'VeryLow' if(51 <= farevalue <= 100) 
            else 'Lower' if (101 <= farevalue <= 150) 
            else 'Low' if (151 <= farevalue <= 200)
            else 'Normal' if (201 <= farevalue <= 250)
            else 'Lower' if (251 <= farevalue <= 300)
            else 'Lower' if (301 <= farevalue <= 350)
            else 'High' if (351 <= farevalue <= 400)
            else 'Higher' if (451 <= farevalue<= 450)
            else 'VeryHigh' if (451 <= farevalue <= 500)
            else 'ExtremeHight')
            

dfcombined['FareClass'] = dfcombined['Fare'].map(binsfunct)
```

```python
fare_grouped = dfcombined.groupby(['FareClass'])
faregroupstats = fare_grouped.describe()
functions = ['count']
result = fare_grouped['Fare'].agg(functions)
result.sort_values(by = 'count', ascending=False)
```

```python
#Lets show this by graph
plt.close('all')
fig = plt.figure()
dfcombined.FareClass.value_counts().plot(kind="barh")
plt.title("FareClass Distribution")
plt.show()
```
